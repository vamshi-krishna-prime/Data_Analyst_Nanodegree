### <ins>0.5 Dataset:</ins>

Dataset Link: **[bikeshare.metro.net](https://bikeshare.metro.net/about/data/)**

This data set includes information about individual rides made in a bike-sharing system covering the `Los Angeles` area.

Each trip is anonymized and includes:

+ `trip_id`: Locally unique integer that identifies the trip
+ `duration`: Length of trip in minutes
+ `start_time`: The date/time when the trip began, presented in ISO 8601 format in local time
+ `end_time`: The date/time when the trip ended, presented in ISO 8601 format in local time
+ `start_station`: The station ID where the trip originated (for station name and more information on each station see the Station Table)
+ `start_lat`: The latitude of the station where the trip originated
+ `start_lon`: The longitude of the station where the trip originated
+ `end_station`: The station ID where the trip terminated (for station name and more information on each station see the Station Table)
+ `end_lat`: The latitude of the station where the trip terminated
+ `end_lon`: The longitude of the station where the trip terminated
+ `bike_id`:  Locally unique integer that identifies the bike
+ `plan_duration`: The number of days that the plan the passholder is using entitles them to ride; 0 is used for a single ride plan (Walk-up)
+ `trip_route_category`: "Round Trip" for trips starting and ending at the same station or "One Way" for all other trips
+ `passholder_type`: The name of the passholder's plan
+ `bike_type`: The kind of bike used on the trip, including standard pedal-powered bikes, electric assist bikes, or smart bikes.

### <ins>1.1 Gather data from `web`:</ins>

```python
# load libraries
import re
import os
import requests
import urllib.request
import numpy as np
import pandas as pd
from io import BytesIO
from zipfile import ZipFile
import glob
import time
import json
import pickle
import math
```

---------------

```python
def extract_filename(url, company):
    year_regex = re.compile(r"(\d{4})-([q]\d)|([Q]\d)_(\d{4})")
    matching_object = year_regex.search(url)
    if matching_object.group(1) == None:
        year = "{}_{}".format(matching_object.group(4), matching_object.group(3)).replace('Q', 'q')
    else:
        year = "{}_{}".format(matching_object.group(1), matching_object.group(2))
    return company + '_' + year


def get_response(url, filename, attempt=1, max_attempts=5):
    '''
    The default User-Agent of requests is python-requests, and in some cases websites don't like traffic 
    from "non-browsers", so they try to block such traffic. This can be avoided by making the request appear 
    like coming from a browser, by adding an extra header parameter.
    '''
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko)\
                Chrome/56.0.2924.76 Safari/537.36'}
    # add time gap between every request to avoid timeout error
    time.sleep(5)
    response = requests.get(url, headers=headers, stream=True, allow_redirects=False, timeout=10)
    try:
        requests.post(url, headers, timeout=10)
    except requests.exceptions.Timeout:
        print("Timeout occurred")
        if attempt <= max_attempts:
            return get_response(url, filename, attempt=attempt+1)
    # check the status of the request
    try:
        response.raise_for_status()
        print('{}'.format(filename).ljust(20, ' '), ':', 'Data request successful.')
    except Exception as exc:
        print('{}'.format(filename).ljust(20, ' '), ':', 'WARNING - "{}"'.format(exc))
        if attempt <= max_attempts:
            return get_response(url, filename, attempt=attempt+1)
    return response


def extract_csv(response):
    try:
        zf = ZipFile(BytesIO(response.content))
        # find the first matching csv file in the zip:
        match = [s for s in zf.namelist() if ".csv" in s][0]
        print('CSV file obtained'.ljust(20,' '), ':', match)
        df = pd.read_csv(zf.open(match), low_memory=False)
        return df
    except BadZipFile:
        print('BadZipFile')


def save_csv_file(df, filename):
    csv_filename = '{}.csv'.format(filename)
    # search for any files which match the filename
    files_present = glob.glob(csv_filename)
    # if no matching files exist, write to csv. else, print warning
    if not files_present:
        df.to_csv(csv_filename, sep=',', index= False)
        print('Action performed'.ljust(20, ' '), ': CSV file saved')
    else:
        print('Warning'.ljust(20, ' '), ': This file already exists!!')


def add_to_dictionary(bikeshare, df, filename, company):
    # add company name to the dataframe (optional)
    # df.insert(0, 'company', company)
    # add dataframe to the dictionary
    bikeshare[str(filename)] = df
    print('Dataset obtained'.ljust(20, ' '), ':', filename)
```

```python
bikeshare = {}
company = 'Metrobike'

print('\n', ' START OF DATA GATHERING '.center(60, '='), '\n')
with open('url_list.txt', encoding='utf-8') as file:
    for line in file:
        url = line.rstrip("\n")
        print('-'*60)
        filename = extract_filename(url, company)
        response = get_response(url, filename)
        df = extract_csv(response)
        # optional to save individual datasets
        # save_csv_file(df, filename)
        add_to_dictionary(bikeshare, df, filename, company)
        print('-'*60)
print('\n')
print(' END OF DATA GATHERING '.center(60, '='))
print('\n')
print(' Check "bikeshare" dictionary for all datasets '.center(60, '/'))
```

```python
def visual_assess(df, col):
    print('{} head'.format(col).center(30, '-'))
    print(df[col].head(), '\n\n')
    print('{} tail'.format(col).center(30, '-'))
    print(df[col].tail(), '\n\n')
    print('{} sample'.format(col).center(30, '-'))
    print(df[col].sample(5), '\n\n')
```

```python
def programmatic_assess(df, col):
    datatype = df[col].dtype
    null_count = df[col].isnull().sum()
    duplicate_count = df[col].duplicated().sum()
    print('{} column'.format(col).center(30, '-'))
    print('datatype'.ljust(15, ' '), ':', datatype)
    print('null_count'.ljust(15, ' '), ':', null_count)
    print('duplicate_count'.ljust(15, ' '), ':', duplicate_count)
```

### <ins>3.6 Feature Engineering</ins>

#### 3.6.1.a Define:

Extract `price` estimation for the bike rental based on `pass_type` and `duration_min`.

Bike rental fare based on pass type excluding the one time pass charges.

| Pass Type   | First 30 min fare | After 30 min fare|
| :---:       |       :----:      |       :---:      |
| Walk-up     |      1 dollars    |   1.75 dollars   |
| One Day     |      0 dollars    |   1.75 dollars   |
| Monthly     |      0 dollars    |   1.75 dollars   |
| Flex        |      0 dollars    |   1.75 dollars   |
| Annual      |      0 dollars    |   1.75 dollars   |

#### 3.6.1.b Code:

```python
%%time

def estimate_fare(df):
    df.loc[df['passholder_type'] == 'Walk-up','fare'] = df.query(' passholder_type == "Walk-up" ')['duration'].apply(lambda x: np.add(np.multiply(np.maximum(0, (math.ceil(x/30)-1)), 1.75),1))
    df.loc[df['passholder_type'] == 'One Day','fare'] = df.query(' passholder_type == "One Day" ')['duration'].apply(lambda x: np.add(np.multiply(np.maximum(0, (math.ceil(x/30)-1)), 1.75),0))
    df.loc[df['passholder_type'] == 'Monthly','fare'] = df.query(' passholder_type == "Monthly" ')['duration'].apply(lambda x: np.add(np.multiply(np.maximum(0, (math.ceil(x/30)-1)), 1.75),0))
    df.loc[df['passholder_type'] == 'Flex','fare'] = df.query(' passholder_type == "Flex" ')['duration'].apply(lambda x: np.add(np.multiply(np.maximum(0, (math.ceil(x/30)-1)), 1.75),0))
    df.loc[df['passholder_type'] == 'Annual','fare'] = df.query(' passholder_type == "Annual" ')['duration'].apply(lambda x: np.add(np.multiply(np.maximum(0, (math.ceil(x/30)-1)), 1.75),0))
    
df = bikeshare_clean
estimate_fare(df)
df.fare.describe()
```

> The customers can be categorized by the fare as `Base` fare for renting the bike under `30 minutes` and `Extended` fare for using the bike more than `30 minutes` and subjected to additional payment.

```python
# Display the base fare rentals based on passholder_type
walkup_basefare = bikeshare_clean.query(' passholder_type == "Walk-up" and fare == 1 ').shape[0]
oneday_basefare = bikeshare_clean.query(' passholder_type == "One Day" and fare == 0 ').shape[0]
monthly_basefare = bikeshare_clean.query(' passholder_type == "Monthly" and fare == 0 ').shape[0]
flex_basefare = bikeshare_clean.query(' passholder_type == "Flex" and fare == 0 ').shape[0]
annual_basefare = bikeshare_clean.query(' passholder_type == "Annual" and fare == 0 ').shape[0]
total_basefare = walkup_basefare + oneday_basefare + monthly_basefare + flex_basefare + annual_basefare

print('total_basefare'.ljust(18, ' '), ':', total_basefare)
print('walkup_basefare'.ljust(18, ' '), ':', walkup_basefare)
print('oneday_basefare'.ljust(18, ' '), ':', oneday_basefare)
print('monthly_basefare'.ljust(18, ' '), ':', monthly_basefare)
print('flex_basefare'.ljust(18, ' '), ':', flex_basefare)
print('annual_basefare'.ljust(18, ' '), ':', annual_basefare)
```

> Extract the `fare_type` based on `passholder_type` and bike rental `duration`.

```python
%%time

def extract_faretype(df):
    df.loc[df.query(' passholder_type == "Walk-up" and fare == 1 ').index,'fare_type'] = "Base"
    df.loc[df.query(' passholder_type == "Walk-up" and fare != 1 ').index,'fare_type'] = "Extended"
    df.loc[df.query(' passholder_type == "One Day" and fare == 0 ').index,'fare_type'] = "Base"
    df.loc[df.query(' passholder_type == "One Day" and fare != 0 ').index,'fare_type'] = "Extended"
    df.loc[df.query(' passholder_type == "Monthly" and fare == 0 ').index,'fare_type'] = "Base"
    df.loc[df.query(' passholder_type == "Monthly" and fare != 0 ').index,'fare_type'] = "Extended"
    df.loc[df.query(' passholder_type == "Flex" and fare == 0 ').index,'fare_type'] = "Base"
    df.loc[df.query(' passholder_type == "Flex" and fare != 0 ').index,'fare_type'] = "Extended"
    df.loc[df.query(' passholder_type == "Annual" and fare == 0 ').index,'fare_type'] = "Base"
    df.loc[df.query(' passholder_type == "Annual" and fare != 0 ').index,'fare_type'] = "Extended"
    
df = bikeshare_clean
extract_faretype(df)
df.fare_type.value_counts()
```
